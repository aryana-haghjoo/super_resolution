```python
import pandas as pd
import glob
from astropy.io import fits

# Path to your FITS files
fits_files = glob.glob("/Users/aryanahaghjoo/Documents/GitHub/super_resolution/data/co_1dspec/*.fits")

# Initialize an empty list to store extracted data
data_list = []

# Loop through each FITS file
for file in fits_files:
    with fits.open(file) as hdul:
        try:
            # Extract object name
            header = hdul[0].header
            object_name = header.get('OBJECT', file)  # Use filename if OBJECT is missing

            # Extract relevant arrays from the correct HDUs
            wavelength = hdul[1].data  # Assuming HDU[1] is Wavelength
            flux = hdul[2].data  # Assuming HDU[2] is Flux
            flux_error = hdul[3].data  # Assuming HDU[3] is Flux Error

            # Store data as arrays instead of individual numbers
            data_list.append([object_name, wavelength, flux, flux_error])

        except Exception as e:
            print(f"Error processing {file}: {e}")

# Convert to Pandas DataFrame
df = pd.DataFrame(data_list, columns=['Object', 'Wavelength', 'Flux', 'Flux_Error'])
```
